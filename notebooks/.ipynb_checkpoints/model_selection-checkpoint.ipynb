```python
#Сделал локальную метрику для первичного анализа моделей
def compute_hit_k(question_embs, corpus_embs, ground_truth, k=3): 
    hits = 0
    for i in range(len(question_embs)):
        sims = util.cos_sim(question_embs[i], corpus_embs)[0]
        top_k_idx = torch.topk(sims, k=k).indices.cpu().numpy()
        if ground_truth[i] in top_k_idx:
            hits += 1
    return hits / len(question_embs)
```
